# Data Preparation

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import label_binarize

df = sns.load_dataset('iris')
X = df.drop('species', axis=1)
y = df['species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Model Building and Evaluation

# K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)

knn_accuracy = accuracy_score(y_test, y_pred_knn)
knn_conf_matrix = confusion_matrix(y_test, y_pred_knn)

y_test_bin = label_binarize(y_test, classes=['setosa', 'versicolor', 'virginica'])
y_pred_prob_knn = knn.predict_proba(X_test)

knn_fpr, knn_tpr, _ = roc_curve(y_test_bin[:, 1], y_pred_prob_knn[:, 1])
knn_auc = auc(knn_fpr, knn_tpr)

print(f"KNN Accuracy: {knn_accuracy:.2f}")
```

# Support Vector Machine (SVM)

```python
from sklearn.svm import SVC

svm = SVC(probability=True, kernel='linear')
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)

svm_accuracy = accuracy_score(y_test, y_pred_svm)
svm_conf_matrix = confusion_matrix(y_test, y_pred_svm)

y_test_bin = label_binarize(y_test, classes=['setosa', 'versicolor', 'virginica'])
y_pred_prob_svm = svm.predict_proba(X_test)

svm_fpr, svm_tpr, _ = roc_curve(y_test_bin[:, 1], y_pred_prob_svm[:, 1])
svm_auc = auc(svm_fpr, svm_tpr)

print(f"SVM Accuracy: {svm_accuracy:.2f}")
```

# Confusion Matrices

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.heatmap(knn_conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=['Setosa', 'Versicolor', 'Virginica'], yticklabels=['Setosa', 'Versicolor', 'Virginica'])
plt.title("KNN Confusion Matrix")
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 2)
sns.heatmap(svm_conf_matrix, annot=True, fmt="d", cmap="Greens", xticklabels=['Setosa', 'Versicolor', 'Virginica'], yticklabels=['Setosa', 'Versicolor', 'Virginica'])
plt.title("SVM Confusion Matrix")

plt.tight_layout()
plt.show()
```

# ROC Curve Comparison: KNN vs SVM

```python
plt.figure(figsize=(10, 8))
plt.plot(knn_fpr, knn_tpr, color='blue', lw=2, label=f'KNN (AUC = {knn_auc:.2f})')
plt.plot(svm_fpr, svm_tpr, color='green', lw=2, label=f'SVM (AUC = {svm_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison: KNN vs SVM')
plt.legend(loc='lower right')
plt.show()

print(f"KNN Accuracy: {knn_accuracy:.2f}")
print(f"SVM Accuracy: {svm_accuracy:.2f}")
```

# Model Comparison

```python
print(f"KNN Accuracy: {knn_accuracy:.2f}")
print(f"SVM Accuracy: {svm_accuracy:.2f}")

model_comparison = pd.DataFrame({
    "Model": ["KNN", "SVM"],
    "Accuracy": [knn_accuracy, svm_accuracy],
    "AUC": [knn_auc, svm_auc]
})
print(model_comparison)
```
